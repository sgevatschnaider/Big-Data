```python
from IPython.display import display, HTML

html_content = """
<!DOCTYPE html>
<html lang="es">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Lightning Network - Clase de Big Data</title>
    <style>
        body {
            font-family: 'Georgia', serif;
            line-height: 1.6;
            margin: 0;
            padding-top: 60px;
            background-color: #f9f9f9;
            background-image: url('https://www.transparenttextures.com/patterns/light-paper-fibers.png');
            background-attachment: fixed;
            padding: 20px;
        }
        .navbar {
            background-color: #0066cc;
            color: white;
            padding: 15px;
            text-align: center;
            font-size: 1.2em;
            position: fixed;
            width: 100%;
            top: 0;
            left: 0;
            z-index: 1000;
            box-shadow: 0px 2px 4px rgba(0, 0, 0, 0.1);
        }
        h1, h2, h3 {
            color: #333;
        }
        h1 {
            text-align: center;
            font-size: 2.5em;
            margin-bottom: 20px;
            font-family: 'Arial', sans-serif;
        }
        h2 {
            color: #0056b3;
            font-size: 1.8em;
            margin-top: 20px;
            cursor: pointer;
            background-color: #e0f7fa;
            padding: 10px;
            border-radius: 5px;
            position: relative;
            transition: background-color 0.3s ease;
        }
        h2::after {
            content: '\\25BC'; /* Triángulo hacia abajo */
            position: absolute;
            right: 20px;
            transition: transform 0.3s ease;
        }
        h2[aria-expanded="true"]::after {
            transform: rotate(180deg); /* Triángulo hacia arriba */
        }
        h3 {
            font-size: 1.5em;
            margin-top: 15px;
            cursor: pointer;
            background-color: #e8f5e9;
            padding: 8px;
            border-radius: 5px;
            position: relative;
            transition: background-color 0.3s ease;
        }
        h3::after {
            content: '\\25BC'; /* Triángulo hacia abajo */
            position: absolute;
            right: 20px;
            transition: transform 0.3s ease;
        }
        h3[aria-expanded="true"]::after {
            transform: rotate(180deg); /* Triángulo hacia arriba */
        }
        p, li {
            font-size: 1.1em;
            margin-bottom: 10px;
        }
        ul {
            list-style-type: none;
            padding-left: 20px;
            display: none;
        }
        ul.visible {
            display: block;
        }
        li {
            margin-bottom: 10px;
        }
        /* Responsividad */
        @media (max-width: 768px) {
            body {
                padding: 10px;
            }
            h1 {
                font-size: 2em;
            }
            .navbar {
                font-size: 1em;
                padding: 10px;
            }
            h2 {
                font-size: 1.5em;
                padding: 8px;
            }
            h3 {
                font-size: 1.3em;
                padding: 6px;
            }
            p, li {
                font-size: 1em;
            }
        }
    </style>
</head>
<body>
    <div class="navbar">
        Material elaborado por Sergio Gevatschnaider, Universidad de Palermo
    </div>

    <h1>La Lightning Network como solución de segunda capa para blockchain</h1>

    <!-- Sección 1 -->
    <h2 onclick="toggleContent('content1')" aria-expanded="false" aria-controls="content1">1.1 Introducción a la Lightning Network</h2>
    <ul id="content1">
        <li>La Lightning Network es un protocolo de segunda capa construido sobre la cadena de bloques de Bitcoin. Fue desarrollado para abordar los problemas de escalabilidad de Bitcoin, específicamente su rendimiento limitado de transacciones y sus altas tarifas durante períodos de alta actividad en la red. La Lightning Network permite que las transacciones ocurran fuera de la cadena; esto significa que las transacciones no se registran inmediatamente en la cadena de bloques principal de Bitcoin. Así, se logran transacciones más rápidas a un costo significativamente menor, lo que mejora la utilidad de Bitcoin como moneda para transacciones diarias y micropagos.</li>
    </ul>

    <h2 onclick="toggleContent('content2')" aria-expanded="false" aria-controls="content2">1.2 ¿Qué es la Lightning Network y cómo funciona en Bitcoin?</h2>
    <ul id="content2">
        <li>La Lightning Network permite a los usuarios crear canales de pago entre sí. Estos canales facilitan que múltiples transacciones se realicen fuera de la cadena de bloques principal, reduciendo la carga en la red de Bitcoin. Así es como funciona:</li>
        <ul>
            <li><b>Apertura de un canal:</b> Dos participantes bloquean una cierta cantidad de Bitcoin en una billetera multifirma, creando un canal; esta transacción inicial se registra en la cadena de bloques principal.</li>
            <li><b>Transacciones fuera de la cadena:</b> Una vez establecido el canal, los participantes pueden realizar transacciones instantáneas entre sí tantas veces como deseen, actualizando cada uno sus respectivos saldos dentro del canal.</li>
            <li><b>Cierre del canal:</b> Cuando los participantes desean finalizar, cierran el canal y el saldo final se registra en la cadena de bloques de Bitcoin.</li>
        </ul>
        <li>La Lightning Network esencialmente permite que Bitcoin crezca procesando transacciones fuera de la cadena, liquidando solo los saldos finales en la cadena de bloques principal. Este enfoque preserva la descentralización y la seguridad de Bitcoin, al tiempo que mejora la velocidad de las transacciones y la eficiencia en los costos.</li>
    </ul>

    <h2 onclick="toggleContent('content3')" aria-expanded="false" aria-controls="content3">1.3 Ventajas de la Lightning Network</h2>
    <ul id="content3">
        <li>
            <h3 onclick="toggleContent('subcontent1')" aria-expanded="false" aria-controls="subcontent1">1.3.1 Velocidad de transacción</h3>
            <ul id="subcontent1">
                <li><b>Transacciones instantáneas:</b> Las transacciones en la Lightning Network se procesan casi instantáneamente, lo que permite utilizar Bitcoin en aplicaciones en tiempo real, como compras minoristas o pagos en línea.</li>
                <li><b>Sin esperar confirmaciones de bloques:</b> A diferencia de las transacciones de Bitcoin en la cadena, que requieren confirmación en un nuevo bloque, las transacciones de Lightning se confirman fuera de la cadena dentro del canal establecido. Esto mejora en gran medida la experiencia del usuario, especialmente en situaciones en las que los pagos urgentes son esenciales.</li>
            </ul>
        </li>
        <li>
            <h3 onclick="toggleContent('subcontent2')" aria-expanded="false" aria-controls="subcontent2">1.3.2 Bajos costos</h3>
            <ul id="subcontent2">
                <li><b>Tarifas de transacción reducidas:</b> Las transacciones en la Lightning Network generalmente incurren en tarifas muy bajas en comparación con las transacciones en la cadena de Bitcoin. Esto se debe a que las transacciones se realizan fuera de la cadena, evitando las altas tarifas de minería que ocurren cuando la red de Bitcoin está congestionada.</li>
                <li><b>Rentabilidad para pagos pequeños:</b> La estructura de bajo costo de la Lightning Network hace factible que los usuarios envíen micropagos, que serían prohibitivamente caros en la cadena de bloques principal de Bitcoin.</li>
            </ul>
        </li>
        <li>
            <h3 onclick="toggleContent('subcontent3')" aria-expanded="false" aria-controls="subcontent3">1.3.3 Soporte para micropagos</h3>
            <ul id="subcontent3">
                <li><b>Habilitación de microtransacciones:</b> Uno de los principales casos de uso de la Lightning Network es su soporte para micropagos. Las transacciones de pequeño valor, como propinas o pagos de bienes y servicios digitales, suelen ser poco prácticas en la cadena de bloques de Bitcoin debido a las elevadas tarifas de transacción.</li>
                <li><b>Oportunidades para nuevos modelos de negocio:</b> El soporte de la Lightning Network para microtransacciones permite nuevas aplicaciones, como servicios de pago por uso, pagos de contenido pequeño y microdonaciones, ampliando la utilidad de Bitcoin más allá de las transacciones financieras tradicionales.</li>
            </ul>
        </li>
    </ul>

    <h2 onclick="toggleContent('content4')" aria-expanded="false" aria-controls="content4">2. Estructura de la Lightning Network como grafo dinámico</h2>
    <ul id="content4">
        <li>La estructura de la Lightning Network se puede representar como un grafo dinámico compuesto por nodos y enlaces (canales), donde cada conexión entre nodos tiene atributos específicos, como la capacidad de transacción y las tarifas de enrutamiento. Esta estructura de grafo admite transacciones rápidas mediante la optimización de las rutas utilizadas para el enrutamiento de pagos. Debido a la naturaleza descentralizada y en constante cambio de la red, el grafo de la Lightning Network es complejo y adaptativo.</li>
    </ul>

    <h2 onclick="toggleContent('content5')" aria-expanded="false" aria-controls="content5">2.1 Representación de la Lightning Network como un grafo de nodos y canales</h2>
    <ul id="content5">
        <li>En la Lightning Network, la arquitectura se visualiza como un grafo formado por nodos (usuarios o participantes de la red) y enlaces (canales de pago que conectan nodos). Cada elemento de este grafo desempeña un papel específico en la facilitación de un enrutamiento de pagos seguro y eficiente.</li>
    </ul>

    <h2 onclick="toggleContent('content6')" aria-expanded="false" aria-controls="content6">2.1.1 Nodos: Usuarios y operadores de red</h2>
    <ul id="content6">
        <li><b>Nodos como participantes:</b> Cada participante en la Lightning Network, ya sea un usuario individual o un gran operador (por ejemplo, intercambios o procesadores de pagos), está representado por un nodo. Estos nodos son responsables de crear, administrar y mantener canales con otros nodos, permitiendo que las transacciones se enruten a través de ellos.</li>
        <li><b>Papel de los nodos en el enrutamiento de pagos:</b> Los nodos de la Lightning Network cumplen una doble función como puntos finales e intermediarios para los pagos. Si un usuario no tiene un canal de pago directo con su destinatario deseado, la red puede enrutar el pago a través de múltiples nodos intermediarios, lo que permite realizar transacciones con prácticamente cualquier otro nodo de la red.</li>
    </ul>

    <h2 onclick="toggleContent('content7')" aria-expanded="false" aria-controls="content7">2.1.2 Enlaces: Canales de pago con capacidades y costos de transacción</h2>
    <ul id="content7">
        <li><b>Canales como enlaces:</b> En la Lightning Network, las conexiones entre nodos se conocen como canales de pago, representados como enlaces en el grafo. Estos canales permiten que dos nodos realicen múltiples transacciones fuera de la cadena sin transmitir cada transacción a la cadena de bloques principal, lo que reduce las tarifas y aumenta la velocidad.</li>
        <li><b>Capacidad de canales:</b> Cada canal tiene una capacidad específica, que representa la cantidad máxima de Bitcoin que se puede transferir a través de él. Esta capacidad se establece cuando se abre el canal, en función de la cantidad de Bitcoin bloqueada por ambas partes en la transacción de financiación del canal.</li>
        <li><b>Costos de transacción:</b> Los canales también tienen tarifas de enrutamiento asociadas, que son pequeñas tarifas que cobran los nodos por reenviar pagos al siguiente nodo en la ruta. Estas tarifas varían en función de las políticas de los nodos y canales, y sirven como incentivo para que los operadores de nodos proporcionen liquidez y mantengan los canales activos.</li>
    </ul>

    <h2 onclick="toggleContent('content8')" aria-expanded="false" aria-controls="content8">2.2 Características del grafo dinámico de la Lightning Network</h2>
    <ul id="content8">
        <li>La estructura de grafo de la Lightning Network no es estática; se adapta y cambia constantemente en respuesta a la actividad de la red. Esta cualidad dinámica presenta desafíos y oportunidades únicos para el enrutamiento de pagos, la gestión de la liquidez y la escalabilidad dentro de la red.</li>
    </ul>

    <h2 onclick="toggleContent('content9')" aria-expanded="false" aria-controls="content9">2.2.1 Cambios en tiempo real en la liquidez</h2>
    <ul id="content9">
        <li><b>Liquidez dinámica de los canales:</b> A medida que los usuarios realizan transacciones, la liquidez dentro de cada canal de pago cambia en tiempo real. Cada transacción actualiza el saldo dentro del canal, lo que afecta la cantidad que se puede enviar o recibir en las transacciones posteriores.</li>
        <li><b>Impacto en el enrutamiento de pagos:</b> Debido a que la liquidez de los canales cambia constantemente, los algoritmos de enrutamiento deben tener en cuenta estos cambios al identificar las rutas que pueden completar con éxito un pago. Los sistemas de enrutamiento pueden utilizar datos en tiempo real o modelos probabilísticos para estimar qué canales tienen más probabilidades de tener suficiente liquidez en un momento dado.</li>
    </ul>

    <h2 onclick="toggleContent('content10')" aria-expanded="false" aria-controls="content10">2.2.2 Importancia de la capacidad y la disponibilidad del canal para el enrutamiento de pagos</h2>
    <ul id="content10">
        <li><b>La capacidad como factor clave:</b> La capacidad de cada canal determina si puede manejar un tamaño de pago en particular. Es posible que los canales de baja capacidad no puedan enrutar pagos grandes, lo que requiere el uso de canales alternativos o la división de pagos en múltiples canales (pagos de múltiples rutas).</li>
        <li><b>Disponibilidad de canales y conectividad de red:</b> La disponibilidad de canales entre nodos afecta la conectividad general de la red e influye en la eficiencia del enrutamiento de pagos. Los nodos con alta disponibilidad y múltiples conexiones a otros nodos funcionan como centros, proporcionando vías sólidas que mejoran las tasas de éxito de los pagos y reducen la latencia. Garantizar una capacidad y conectividad de canal adecuadas es esencial para mantener una red de pago fluida y escalable dentro del grafo dinámico de la Lightning Network.</li>
    </ul>

    <h2 onclick="toggleContent('content11')" aria-expanded="false" aria-controls="content11">3. Algoritmos de enrutamiento en la Lightning Network</h2>
    <ul id="content11">
        <li>Descripción general de los algoritmos utilizados para optimizar el enrutamiento de pagos en la Lightning Network.</li>
    </ul>

    <h2 onclick="toggleContent('content12')" aria-expanded="false" aria-controls="content12">3.2 Algoritmo de Dijkstra modificado</h2>
    <ul id="content12">
        <li>El algoritmo de Dijkstra modificado es una adaptación del algoritmo de Dijkstra tradicional, diseñado para manejar los desafíos específicos de la Lightning Network, donde cada canal tiene restricciones específicas de capacidad y tarifa.</li>
    </ul>

    <h2 onclick="toggleContent('content13')" aria-expanded="false" aria-controls="content13">3.2.1 Adaptación de Dijkstra para redes con limitaciones de capacidad y costo</h2>
    <ul id="content13">
        <li><b>Requisito de capacidad mínima y optimización de costos:</b> A diferencia del Dijkstra estándar, que solo encuentra el camino más corto en función de la distancia, la versión modificada considera tanto la capacidad del canal como las tarifas de enrutamiento. El algoritmo selecciona solo los canales que cumplen con el requisito de capacidad mínima para el pago e intenta encontrar la ruta con el costo acumulado más bajo.</li>
    </ul>

    <h2 onclick="toggleContent('content14')" aria-expanded="false" aria-controls="content14">3.2.2 Pasos detallados del algoritmo de Dijkstra modificado en la Lightning Network</h2>
    <ul id="content14">
        <li><b>Inicializar costos y capacidad:</b> Comienza con el nodo de origen e inicializa todas las distancias hasta el infinito, excepto el nodo de origen (establecido en cero). Se inicializa un requisito de capacidad mínima en función del tamaño del pago.</li>
        <li><b>Seleccionar canales factibles:</b> Para cada nodo, evalúa solo los canales (enlaces) con capacidad suficiente y suma el costo de transacción acumulado.</li>
        <li><b>Actualizar rutas:</b> Para cada ruta viable, actualiza el costo acumulado y la ruta solo si proporciona un costo menor que las rutas calculadas anteriormente.</li>
        <li><b>Terminar al llegar al destino:</b> Cuando se alcanza el nodo de destino con una ruta factible, genera la ruta con el costo acumulado más bajo que cumpla con el requisito de capacidad.</li>
    </ul>

    <h2 onclick="toggleContent('content15')" aria-expanded="false" aria-controls="content15">3.2.3 Ventajas y limitaciones de Dijkstra en la Lightning Network</h2>
    <ul id="content15">
        <li><b>Comparación con Flare y A*:</b>
            <ul>
                <li><b>Requisito de conocimiento global:</b> Dijkstra asume el conocimiento de la topología y los estados del canal de toda la red, lo que puede ser poco práctico en redes altamente dinámicas como Lightning.</li>
                <li><b>Eficiencia:</b> Si bien Dijkstra encuentra el camino más corto con el menor costo, es posible que no se adapte bien a redes donde los estados del canal cambian con frecuencia. A diferencia de Flare, que utiliza datos probabilísticos, Dijkstra requiere datos en tiempo real, lo que puede hacer que sea menos eficiente en entornos dinámicos.</li>
            </ul>
        </li>
    </ul>

    <h2 onclick="toggleContent('content16')" aria-expanded="false" aria-controls="content16">3.3 Búsqueda modificada A* (A-Star) para la Lightning Network</h2>
    <ul id="content16">
        <li>El algoritmo A* es otro algoritmo de búsqueda de rutas que se basa en el enfoque de Dijkstra, pero incorpora una función heurística para priorizar los nodos que están más cerca del destino, haciéndolo más eficiente en redes grandes.</li>
    </ul>

    <h2 onclick="toggleContent('content17')" aria-expanded="false" aria-controls="content17">3.3.1 Descripción del algoritmo A* y su uso de heurística en la Lightning Network</h2>
    <ul id="content17">
        <li><b>Funcionamiento:</b> A* combina el cálculo de la ruta más corta con una función heurística que estima la distancia desde el nodo actual hasta el destino. Esto permite a A* priorizar las rutas que tienen más probabilidades de conducir a la meta, reduciendo el número de nodos que deben explorarse.</li>
        <li><b>Adaptación en Lightning Network:</b> En Lightning Network, la heurística se puede adaptar para considerar factores como la proximidad, las tarifas del canal y la liquidez estimada, lo que proporciona una forma eficiente de enrutar los pagos incluso cuando la información completa de la red no está disponible.</li>
    </ul>

    <h2 onclick="toggleContent('content18')" aria-expanded="false" aria-controls="content18">3.3.2 Comparación de A* con Dijkstra y Flare en redes de pago dinámicas</h2>
    <ul id="content18">
        <li><b>Dijkstra vs. A*:</b> A* puede ser más eficiente que Dijkstra en redes grandes, ya que utiliza una heurística para priorizar los nodos, reduciendo el tiempo de cálculo al ignorar las rutas menos prometedoras.</li>
        <li><b>A* vs. Flare:</b> A diferencia de Flare, que utiliza nodos de probabilidad y baliza, A* se basa en una heurística para la priorización. Si bien es efectivo, A* puede no ser tan adaptable a la liquidez dinámica y las actualizaciones frecuentes requeridas por la Lightning Network.</li>
    </ul>

    <h2 onclick="toggleContent('content19')" aria-expanded="false" aria-controls="content19">3.3.3 Ventajas y limitaciones de A* en la Lightning Network en relación con:</h2>
    <ul id="content19">
        <li><b>Precisión de ruta:</b> A* puede proporcionar un enrutamiento preciso, especialmente en los casos en que una heurística puede estimar eficazmente la distancia o el costo. Sin embargo, puede ser limitado en un entorno dinámico donde la liquidez y las tarifas en tiempo real fluctúan.</li>
        <li><b>Recursos computacionales:</b> A* es generalmente más eficiente computacionalmente que Dijkstra, especialmente en redes grandes, pero su dependencia de la heurística significa que es posible que no tenga en cuenta los cambios de liquidez en tiempo real con tanta eficacia como Flare.</li>
    </ul>

    <h2 onclick="toggleContent('content20')" aria-expanded="false" aria-controls="content20">4. Conclusión</h2>
    <ul id="content20">
        <li>La Lightning Network representa un avance significativo para la escalabilidad de Bitcoin al permitir transacciones rápidas, de bajo costo y seguras a través de una solución de segunda capa. Como red fuera de la cadena, permite que las transacciones se realicen fuera de la cadena de bloques principal, lo que reduce la congestión de la red y reduce las tarifas de transacción. Esto hace posible el uso de Bitcoin para los pagos cotidianos, incluidos los micropagos, que de otro modo serían prohibitivamente caros en la cadena de bloques principal.</li>
        <li>Estructuralmente, la Lightning Network funciona como un grafo dinámico formado por nodos (usuarios y operadores de red) y enlaces (canales de pago). Este grafo está en constante evolución a medida que los nodos abren y cierran canales o ajustan su liquidez, lo que lo hace potente y complejo. Cada canal tiene una capacidad específica y un costo de transacción asociado, que afectan directamente las decisiones de enrutamiento de pagos. La adaptabilidad de la red permite ajustes de enrutamiento en tiempo real, pero también presenta desafíos, particularmente para mantener una liquidez adecuada y un enrutamiento eficiente a través de una topología en constante cambio.</li>
        <li>Para hacer frente a estos desafíos, la Lightning Network se basa en algoritmos de enrutamiento avanzados diseñados para encontrar rutas de pago eficientes. Algoritmos como Flare, Dijkstra modificado y A* aportan fortalezas únicas. Flare, con su uso de exploración local y nodos de baliza, aprovecha el enrutamiento basado en probabilidades para optimizar el éxito de los pagos en una red dinámica. El algoritmo de Dijkstra modificado se adapta a las limitaciones de capacidad y costo, aunque requiere un amplio conocimiento de la red y puede ser menos eficiente en entornos altamente dinámicos. Mientras tanto, A* mejora el enfoque de Dijkstra mediante el uso de heurística para priorizar caminos prometedores, lo que lo hace computacionalmente eficiente en grandes redes, aunque menos adaptable a cambios rápidos de liquidez.</li>
    </ul>

    <script>
        function toggleContent(id) {
            var content = document.getElementById(id);
            var header = document.querySelector(`h2[aria-controls="${id}"], h3[aria-controls="${id}"]`);
            if (content.classList.contains('visible')) {
                content.classList.remove('visible');
                header.setAttribute('aria-expanded', 'false');
            } else {
                content.classList.add('visible');
                header.setAttribute('aria-expanded', 'true');
            }
        }
        // Inicialización: Ocultar todas las secciones
        document.querySelectorAll('ul').forEach(ul => ul.classList.remove('visible'));
    </script>

</body>
</html>
"""

# Mostrar el contenido HTML corregido en Google Colab
display(HTML(html_content))
```
